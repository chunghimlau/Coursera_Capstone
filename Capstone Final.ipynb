```python
# Drop all columns with no predictive value for the context of this project
#df_1 = df[['SEVERITYCODE','WEATHER','ROADCOND','LIGHTCOND','PEDCYLCOUNT','UNDERINFL']]


df_1 = df.drop(columns = ['OBJECTID', 'SEVERITYCODE.1', 'REPORTNO', 'INCKEY', 'COLDETKEY', 
              'X', 'Y', 'STATUS','ADDRTYPE',
              'INTKEY', 'LOCATION', 'EXCEPTRSNCODE',
              'EXCEPTRSNDESC', 'SEVERITYDESC', 'INCDATE',
              'INCDTTM', 'JUNCTIONTYPE', 'SDOT_COLCODE',
              'SDOT_COLDESC', 'PEDROWNOTGRNT', 'SDOTCOLNUM',
              'ST_COLCODE', 'ST_COLDESC', 'SEGLANEKEY',
              'CROSSWALKKEY', 'HITPARKEDCAR', 'PEDCOUNT', 
              'PERSONCOUNT', 'VEHCOUNT', 'COLLISIONTYPE',
              'SPEEDING', 'INATTENTIONIND'])

# Label Encoding
# Convert column to category
df_1["WEATHER"] = df_1["WEATHER"].astype('category')
df_1["ROADCOND"] = df_1["ROADCOND"].astype('category')
df_1["LIGHTCOND"] = df_1["LIGHTCOND"].astype('category')

# Assign variable to new column for analysis
df_1["WEATHER_CODE"] = df_1["WEATHER"].cat.codes
df_1["ROADCOND_CODE"] = df_1["ROADCOND"].cat.codes
df_1["LIGHTCOND_CODE"] = df_1["LIGHTCOND"].cat.codes

df_1.head(5)
```

```python
#Showing the unique value counts of the LIGHT CONDITION
df_1['LIGHTCOND'].value_counts()
```

```python
from sklearn.utils import resample
# Splitting majority and minority classes
df_1_majority = df_1[df_1.SEVERITYCODE==1]
df_1_minority = df_1[df_1.SEVERITYCODE==2]

#Turning majority class to Downsample majority class
df_1_majority_downsampled = resample(df_1_majority,
                                        replace=False,
                                        n_samples=58188,
                                        random_state=123)

# Combine minority class with Downsampled majority class
df_1_balanced = pd.concat([df_1_majority_downsampled, df_1_minority])

# Showing the unique value counts of the WEATHER
df_1_balanced.SEVERITYCODE.value_counts()
```

```python
import numpy as np
X = np.asarray(df_1_balanced[['WEATHER_CODE', 'ROADCOND_CODE', 'LIGHTCOND_CODE']])
X[0:5]
```

```python
y = np.asarray(df_1_balanced['SEVERITYCODE'])
y [0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# Building the KNN Model
from sklearn.neighbors import KNeighborsClassifier

k = 25
```

```python
#Train Model & Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh

Kyhat = neigh.predict(X_test)
Kyhat[0:5]
```

```python
# Train Model & Predicr
LRyhat = LR_model.predict(X_test)
LRyhat
```
